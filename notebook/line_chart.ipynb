```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

Datasetin oxunması və dataframe hazırlanması.

```python
df = pd.read_csv(r"C:\Users\User\Desktop\GitHub\DataVisualization\Line_chart\data\line_chart_data.csv")
```

İnfoda date column formatı adi object formatıdır. Onu date formatına çevirək (pandasla)

```python
df['New_Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
sns.set_style('white')
```

```python
labels = ['FTSE 100', 'SAP 500']
plt.figure(figsize=(20, 8))
plt.plot(df['New_Date'], df['FTSE100'])
plt.plot(df['New_Date'], df['GSPC500'])
plt.title("SAP & FTSE returns (2000-2010)", fontsize = 15, weight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

Qrafik uzun illəri əhatə edir. Spesifik olaraq 2009cu ilin 1-ci yarımilinə baxaq.

```python
df_H1_2009 = df[(df['New_Date']>='2009-01-01') &
            (df['New_Date']<='2009-06-30')]
```

```python
labels = ['FTSE 100', 'SAP 500']
plt.figure(figsize=(20, 8))
plt.plot(df_H1_2009['New_Date'], df_H1_2009['FTSE100'], color = "crimson")
plt.plot(df_H1_2009['New_Date'], df_H1_2009['GSPC500'], color = 'midnightblue')
plt.title("SAP & FTSE returns (2009 1st Half)", fontsize = 15, weight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.savefig("LineChartVizual.png")
plt.show()
```
